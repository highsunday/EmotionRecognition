```python
import os, shutil
from glob import glob
```

```python
test_dir='images/fer2013/test'
val_dir='images/fer2013/valid'
train_dir='images/fer2013/train'
```

```python
original_dataset_dir='images/fer2013/All'
list_1=os.listdir(original_dataset_dir)
list_1
```

```python
for folder in list_1:  
    count=0
    files_num=len(glob(original_dataset_dir + '/'+folder+'/*.jpg'))
    for image in glob(original_dataset_dir + '/'+folder+'/*.jpg'):
        #print(image)
        #print(image.split("/")[-1])
        image_name=image.split("/")[-1]
        src = os.path.join(image)
        if(count<files_num/10):
            dst = os.path.join(test_dir+'/'+folder, image_name)
        elif(count<(files_num/10)*2):
            dst = os.path.join(val_dir+'/'+folder, image_name)
        else:
            dst = os.path.join(train_dir+'/'+folder, image_name)
        count+=1
        #print('src:',src)
        #print('dst:',dst)
        shutil.copyfile(src, dst)
```
